```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential, InteractiveBrowserCredential
from azure.ai.ml.constants import AssetTypes

# Same auth + config you use elsewhere
try:
    credential = DefaultAzureCredential()
    credential.get_token("https://management.azure.com/.default")
except Exception:
    credential = InteractiveBrowserCredential()

ml_client = MLClient.from_config(credential=credential, path="../config.json")

# 1) Check data asset (you already did, but keep here for completeness)
data_asset_name = "diabetes-clinical-enriched-130us"
version = "1124_1331"   # your actual version
data_asset = ml_client.data.get(name=data_asset_name, version=version)
print("DATA OK:", data_asset.id, data_asset.type)

# 2) Check compute exists
cluster_name = "clinical-cluster-cpu"
cluster = ml_client.compute.get(cluster_name)
print("COMPUTE OK:", cluster.name, cluster.type)

# 3) Check environment exists (explicit version)
env_name = "clinical-prep-env"
env_version = "1"   # VERY IMPORTANT: adjust if Studio shows a different version
env = ml_client.environments.get(name=env_name, version=env_version)
print("ENV OK:", env.name, env.version)
```
